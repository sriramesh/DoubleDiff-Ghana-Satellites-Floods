```python
import pandas as pd
from os import listdir
from os.path import isfile, join
import pathlib
```

```python
# file paths and file names
path_main = '/Users/srilakshmi/Desktop/Thesis/Data/National-level/Daily-Precipitation/'
name_folder_data = 'Data-NRT/'
path_data = path_main+name_folder_data
name_output_file = 'National-level_Precipitation_NRT.csv'

```

```python
# get list of file names
files = [f for f in listdir(path_data) if isfile(join(path_data, f))]
files.sort(reverse = False)
files
```

```python
# merge files
for i in range(0, 1):
    x = pd.read_csv(name_folder_data+files[i], header=0)
    y = pd.read_csv(name_folder_data+files[i+1], header=0)
    merged = pd.concat([x, y])

for i in range(2, len(files)):
    z = pd.read_csv(name_folder_data+files[i], header=0)
    merged = pd.concat([merged, z])
    i += 1

merged
```

```python
# save file to disk

merged.to_csv(path_main+name_output_file, index = False, header=True)
```
