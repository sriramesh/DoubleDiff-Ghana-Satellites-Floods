```python
import pandas as pd

path_input = '/Users/srilakshmi/Desktop/Thesis/04-Data-Analysis-Results/00-Data-Sources/00-Cloud-to-Street/01-District-level/Data-Observational/'
name_input_file = 'Ghana_Stats_Rec_v2.csv'

path_output = '/Users/srilakshmi/Desktop/Thesis/04-Data-Analysis-Results/00-Data-Sources/00-Cloud-to-Street/01-District-level/Data-Observational/Output/'
name_output_file = 'Ghana_Stats_Rec_FINAL.csv'

df = pd.read_csv(path_input+name_input_file)
```

```python
path_input = '/Users/srilakshmi/Desktop/Thesis/04-Data-Analysis-Results/00-Data-Sources/00-Cloud-to-Street/01-District-level/Totals/'
name_input_file = 'Impacts-Ghana_MaxFlood_20200715_20201014.csv'

dfx = pd.read_csv(path_input+name_input_file)

totals = dfx[[dfx.columns[1], dfx.columns[2], dfx.columns[7], dfx.columns[9]]]
totals.columns = ['district', 'total_population', 'total_cropland', 'total_flood_area']
totals

pop = totals.dropna(subset=['total_population'])[['district', 'total_population']]
cropland = totals.dropna(subset=['total_cropland'])[['district', 'total_cropland']]
floodarea = totals.dropna(subset=['total_flood_area'])[['district', 'total_flood_area']]

x = pop.merge(cropland, on='district', how='outer')
x = x.merge(floodarea, on='district', how='outer')
x = df.merge(x, left_on='admin2name', right_on='district')

x['perc_pop'] = x['pop_sum']/x['total_population']
x['perc_cropland'] = x['crop_area']/x['total_cropland']
x['perc_floodarea'] = x['flood_area']/x['total_flood_area']

x['perc_pop'] = x['perc_pop'].fillna(0)
x['perc_cropland'] = x['perc_cropland'].fillna(0)
x['perc_floodarea'] = x['perc_floodarea'].fillna(0)
x
```

```python
# write to csv ---

x.to_csv(path_output+name_output_file, index = False, header=True)
```
