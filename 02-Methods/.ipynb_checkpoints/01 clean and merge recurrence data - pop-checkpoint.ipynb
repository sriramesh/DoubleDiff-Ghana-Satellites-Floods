## Methods: 01 Clean and merge recurrence and damage estimates data - population

**The purpose of this notebook is to collate all flooded population estimates data to be in a more usable format.** Specifically, this notebook collates flooded population estimates per each recurrence interval into one dataset.


```python
from os import listdir
from os.path import isfile, join
import pathlib
import sys
import constants # custom py module

# load file paths ---
path_mod = constants.path_c2s_district
sys.path.insert(1, path_mod)

path_input = constants.path_population_recurrence
path_output = constants.path_output

# define column names ---
col_name_old = "People Impacted"

# define character location
recur_left = 42
recur_right = 48

# get list of file names
files = [f for f in listdir(path_input) if isfile(join(path_input, f))]
files.sort(reverse = False)
files
```

```python
import clean

# merge files
for i in range(0, 1):
    x = clean.clean_csv_recurrence(path_input, files[i], col_name_old, recur_left, recur_right)
    y = clean.clean_csv_recurrence(path_input, files[i+1], col_name_old, recur_left, recur_right)
    merged = clean.special_merge(left = x, right = y)
    
for i in range(2, 5):
    z = clean.clean_csv_recurrence(path_input, files[i], col_name_old, recur_left, recur_right)
    merged = clean.special_merge(left = merged, right = z)
    i += 1

merged
```

```python
# save file to disk

filename = constants.name_output_recurrence_population
merged.to_csv(path_output+filename, index = False, header=True)
```
