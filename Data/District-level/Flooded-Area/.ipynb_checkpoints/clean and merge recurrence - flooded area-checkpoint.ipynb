```python
import pandas as pd
from os import listdir
from os.path import isfile, join
import pathlib
```

```python
# file paths and file names
col_name_old = 'Flooded Area Impacted'
path_main = '/Users/srilakshmi/Desktop/Thesis/Data/District-level/Flooded-Area/'
name_folder_data = 'Data-Recurrence/'
path_data = path_main+name_folder_data
name_output_file = 'District-level_Flooded_Area_Recurrence.csv'

```

```python
# data cleaning functions

def clean_csv_recurrence(string):
    
    df = pd.read_csv(name_folder_data+string, header=0)
    df = df.drop_duplicates()
    df = df.rename(columns={col_name_old: string[48:54]})
    df = df.sort_values(by=['Region'])

    return df

def special_merge(left,right):
    merged = left.set_index('Region').join(right.set_index('Region')) # merged file 0 and file 1 together    
    merged.index.name = 'Region'
    merged.reset_index(inplace=True)
    
    return merged
```

```python
# get list of file names
files = [f for f in listdir(path_data) if isfile(join(path_data, f))]
files.sort(reverse = False)
files
```

```python
# merge files
for i in range(0, 1):
    x = clean_csv_recurrence(files[i])
    y = clean_csv_recurrence(files[i+1])
    merged = special_merge(left = x, right = y)
    
for i in range(2, 5):
    z = clean_csv_recurrence(files[i])
    merged = special_merge(left = merged, right = z)
    i += 1

merged
```

```python
# save file to disk

merged.to_csv(path_main+name_output_file, index = False, header=True)
```
