```python
import pandas as pd
from os import listdir
from os.path import isfile, join
import pathlib
```

```python
# file paths and names
path_input = '/Users/srilakshmi/Desktop/Thesis/04-Data-Analysis-Results/00-Data-Sources/00-Cloud-to-Street/01-District-level/People/Output/'
name_input_file = 'District-level_People_Recurrence.csv'

path_output = '/Users/srilakshmi/Desktop/Thesis/04-Data-Analysis-Results/00-Data-Sources/00-Cloud-to-Street/01-District-level/People/Output/'
name_output_file = 'District_Flood_Risk_Classification_People.csv'
```

```python
# add in zeroes column
df = pd.read_csv(path_input+name_input_file, header=0)
zeroes = [0] * len(df)
df.insert(1, "Zeroes", zeroes, True)
df
```

```python
import numpy as np

# calculate average impact per year for population

return_periods = np.array([2, 5, 10, 15, 20])
weights = np.array([0.58, 0.2, 0.1, 0.07, 0.05])
avg_impact_per_year = []

for i in range(0, len(df)):

    marg_impact_per_year = [] # collect marginal impact per year to then sum up

    for j in range(1, 6):
        
        x = ( df.iloc[i, j+1] - df.iloc[i, j] ) / return_periods[j-1]
        marg_impact_per_year.append(x)

    avg_impact_per_year.append(sum(marg_impact_per_year * weights)) # weighted sum based on recurrence intervals
```

```python
# add output to df
flood_risk_classification = df[['Region']]
colname = 'Avg Impact Per Year - People'

flood_risk_classification.insert(1,colname, avg_impact_per_year, True)
flood_risk_classification = flood_risk_classification.sort_values(colname) # sort from low to high avg impact per year
flood_risk_classification
```

```python
# save file to disk

flood_risk_classification.to_csv(path_output+name_output_file, index = False, header=True)
```
