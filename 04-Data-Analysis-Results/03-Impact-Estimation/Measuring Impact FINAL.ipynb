## Measuring Impact

Impact would be measured by double difference between the last observation before the introduction of the scheme (2018 - 2020) and the last observation under the scheme. So this would be:

* Before: Last obs of 2017
* After: Last obs of 2020

ATE = (Avg_T_2020 - Avg_T_2017) - (Avg_C_2020 - Avg_C_2017)


```python
var = 'pop_sum'

import pandas as pd

## output files
path_input = '/Users/srilakshmi/Desktop/Thesis/04-Data-Analysis-Results/02-Parallel-Trends-Analysis/'
name_input_file = 'Before_After_Outcome_Variable_Of_Interest.csv'

x = pd.read_csv(path_input+name_input_file)
x = x[x['year'] == 2017].append(x[x['year'] == 2019]) # only keep before and after data
x.columns = ['region', 'treatment', 'post', var]
x['post'] = x['post'].map({2017:0, 2019:1})
x
```

```python
import statsmodels.formula.api as smf

modelX = smf.ols(formula='pop_sum ~ treatment + post + (treatment * post)', data = x).fit()
print(modelX.summary())
```
