```python
import pandas as pd
from os import listdir
from os.path import isfile, join
import pathlib

```

```python
# file paths and file names
path_input = '/Users/srilakshmi/Desktop/Thesis/Treatment-Control-Assignment/Avg-Num-Impacted-Per-Year-Recurrence/'

path_output = '/Users/srilakshmi/Desktop/Thesis/Avg-Num-Impacted-Per-Year-Recurrence/'
name_output_file = 'District_Avg_Num_Impacted_Per_Year_All.csv'
```

```python
# get list of file names
files = [f for f in listdir(path_input) if isfile(join(path_input, f))]
files
```

```python
# merge files
for i in range(0, 1):
    x = pd.read_csv(path_input+files[i])
    y = pd.read_csv(path_input+files[i+1])
    merged = pd.merge(left=x, right=y, how='left', on='Region') 

for i in range(2, 7):
    z = pd.read_csv(path_input+files[i])
    merged = pd.merge(left=merged, right=z, how='left', on='Region')
merged
```

```python
# organize df

colnames = list(merged.columns[1:])
myorder = [4, 6, 0, 1, 2, 3, 5]
colnames = [colnames[i] for i in myorder]

colnames.insert(0, 'Region')
colnames
```

```python
# sort df
merged = merged[colnames]
merged = merged.sort_values(by = ['Avg Impact Per Year - Pop', 'Avg Impact Per Year - Crops'],
                            ascending = False)
merged
```

```python
# save file to disk

merged.to_csv(path_output+name_output_file, index = False, header=True)
```
